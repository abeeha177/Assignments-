```python
# Question 3
def  calculate_discount(price:float,discount_perc:float=10.0):
    disc_amount=(discount_perc/100)*price
    return disc_amount-price
print(calculate_discount(1000))
print(calculate_discount(1000, 20))
#First call: only price
#Second call: price + discount 
```
